```python
import pandas as pd
import numpy as np
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import HillClimbSearch, BDeuScore, BayesianEstimator
import networkx as nx
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Discretized Data
df = pd.read_csv('../data/processed/data_discretized_final.csv')
print(f"Loaded: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
```

```python
system_features = [
    'Motion_Intensity'
]

physio_features = [
   'HR',
    'GSR',
    'Left_Diameter',
    'Right_Diameter',
    'Left_Openness',
    'Right_Openness',
    'Gaze_Error_Angle',
    'HRV',
    'Eye_Blink_Rate',
]

target = 'fms'

print(f"System features: {len(system_features)}")
print(f"Physio features: {len(physio_features)}")
print(f"Target: {target}")
```

```python
min_samples = df['fms'].value_counts().min() 
df_balanced = df.groupby('fms').apply(
    lambda x: x.sample(min_samples, replace=False, random_state=42)
).reset_index(drop=True)

print(f"Balanced: {df_balanced.shape}") 
print(f"Loaded: {df.shape}")
```

```python
black_list = []

# Physio cannot cause System
for physio in physio_features:
    for system in system_features:
        black_list.append((physio, system))

# FMS cannot cause anything
all_features = system_features + physio_features
for feature in all_features:
    black_list.append((target, feature))

print(f"Black list edges: {len(black_list)}")
print(f"\nExample forbidden edges:")
for edge in black_list[:5]:
    print(f"  {edge[0]} → {edge[1]}: FORBIDDEN")
```

```python
hc = HillClimbSearch(df)
best_model = hc.estimate(
    scoring_method=BDeuScore(df, equivalent_sample_size=5),
    max_indegree=3,
    black_list=black_list,
    show_progress=True
)
print(f"Edges: {len(best_model.edges())}")
print(f"\nFirst 10 edges:")
for edge in list(best_model.edges())[:]:
    print(f"  {edge[0]} → {edge[1]}")
```

```python
# Node Reduction
def get_ancestors(model, node):
    ancestors = set()
    queue = [node]
    while queue:
        current = queue.pop(0)
        for parent in model.get_parents(current):
            if parent not in ancestors:
                ancestors.add(parent)
                queue.append(parent)
    return ancestors

ancestors_of_fms = get_ancestors(best_model, target)
nodes_to_keep = list(ancestors_of_fms) + [target]

print(f"Ancestors of FMS: {len(ancestors_of_fms)}")
print(f"Nodes to keep: {len(nodes_to_keep)}")
print(f"\nReduced nodes:")
for node in sorted(nodes_to_keep):
    print(f"  - {node}")

# Create reduced structure
edges_to_keep = [e for e in best_model.edges() if e[0] in nodes_to_keep and e[1] in nodes_to_keep]
reduced_model = BayesianNetwork(edges_to_keep)

print(f"\n Reduced model:")
print(f"  Nodes: {len(reduced_model.nodes())}")
print(f"  Edges: {len(reduced_model.edges())}")
```

```python
reduced_model.fit(
    df[nodes_to_keep],
    estimator=BayesianEstimator,
    prior_type='BDeu',
    equivalent_sample_size=10
)
print(f"\nCPTs created: {len(reduced_model.get_cpds())}")
```

```python
#Inspect CPTs
print("\nSample CPT (FMS):")
fms_cpd = reduced_model.get_cpds(target)
print(fms_cpd)
assert reduced_model.check_model(), "Model invalid!"
```

```python
# Visualize DAG

plt.figure(figsize=(16, 12))

# Use circular or kamada_kawai layout (more stable than spring)
try:
    pos = nx.kamada_kawai_layout(reduced_model)
except:
    pos = nx.spring_layout(reduced_model, k=2, iterations=50, seed=42)

# Color nodes by type
node_colors = []
for node in reduced_model.nodes():
    if node == target:
        node_colors.append('red')
    elif node in system_features:
        node_colors.append('lightblue')
    else:
        node_colors.append('lightgreen')

# Draw with simpler arrow style (no FancyArrowPatch)
nx.draw_networkx_nodes(reduced_model, pos, 
                       node_color=node_colors,
                       node_size=3000,
                       alpha=0.9)

nx.draw_networkx_labels(reduced_model, pos,
                        font_size=8,
                        font_weight='bold')

# Use simple arrows (this avoids the bug)
nx.draw_networkx_edges(reduced_model, pos,
                       arrows=True,
                       arrowstyle='->',
                       arrowsize=15,
                       edge_color='gray',
                       width=2,
                       connectionstyle='arc3,rad=0.1',  # Curved edges
                       node_size=3000)  # Important: match node size

plt.title('Bayesian Network Structure\n(Blue=System, Green=Physio, Red=Target)', 
          fontsize=14, pad=20)
plt.axis('off')
plt.tight_layout()

# Save
plt.savefig('../data/results/bn_structure_1.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Saved: bn_structure.png")
```

```python
plt.figure(figsize=(20, 14))

# Create hierarchical positions
def hierarchical_layout(G, layers):
    """Create layered layout for hierarchical BN"""
    pos = {}
    layer_nodes = {0: [], 1: [], 2: []}
    
    # Assign nodes to layers
    for node in G.nodes():
        if node == 'fms':
            layer_nodes[2].append(node)
        elif node in system_features:
            layer_nodes[0].append(node)
        else:
            layer_nodes[1].append(node)
    
    # Position nodes
    y_positions = [2.0, 1.0, 0.0]  # Top, middle, bottom
    
    for layer_idx, nodes in layer_nodes.items():
        n_nodes = len(nodes)
        if n_nodes > 0:
            x_positions = [(i - (n_nodes-1)/2) * 0.3 for i in range(n_nodes)]
            for i, node in enumerate(sorted(nodes)):
                pos[node] = (x_positions[i], y_positions[layer_idx])
    
    return pos

# Get hierarchical layout
pos = hierarchical_layout(reduced_model, 
                          {'system': system_features, 
                           'physio': [n for n in reduced_model.nodes() 
                                     if n not in system_features and n != 'fms'],
                           'target': ['fms']})

# Color nodes
node_colors = []
for node in reduced_model.nodes():
    if node == target:
        node_colors.append('red')
    elif node in system_features:
        node_colors.append('lightblue')
    else:
        node_colors.append('lightgreen')

# Draw
nx.draw_networkx_nodes(reduced_model, pos, 
                       node_color=node_colors,
                       node_size=4000,
                       alpha=0.9)

nx.draw_networkx_labels(reduced_model, pos,
                        font_size=9,
                        font_weight='bold')

nx.draw_networkx_edges(reduced_model, pos,
                       arrows=True,
                       arrowstyle='->',
                       arrowsize=20,
                       edge_color='gray',
                       width=2.5,
                       alpha=0.6,
                       node_size=4000)

plt.title('Bayesian Network Structure (Hierarchical Layout)\n' + 
          'Top: System Features | Middle: Physiological | Bottom: FMS',
          fontsize=16, pad=30)
plt.axis('off')
plt.tight_layout()

plt.savefig('../data/results/bn_structure_hierarchical_1.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved: bn_structure_hierarchical.png")
```
